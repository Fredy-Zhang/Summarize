```python
gpu_memory_path = "batch_size_8_memory_usage.log"
```

```python
with open(gpu_memory_path, "r") as f:
    lines = f.readlines()
```

```python
new_lines = [ int(x.replace("MiB\n", "")) for x in lines if "memory.used" not in x ]
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
plt.plot(new_lines)
# plt.axvline(x = 81920, color = 'r', label = 'axvline - full height')
plt.axhline(y=81920, color='r', linestyle='--', label='Vertical Line')
plt.xlabel("Time")
plt.ylabel("MiB")
plt.title("The GPU Usage when the Batch size is 8 with patch size (64,64,64)")
plt.savefig("memory_usage_batch_size_8.png")
```
